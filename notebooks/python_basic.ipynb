# Basic
## loop
Using loops with :An block after a loop executes after the loop finishes normally (i.e., without hitting a statement).

```python
numbers = [1,2,3,4,5]
for num in numbers:
    print(num)
else: 
    print("Loop completed")
```

# Section : Comprehensions

### List Comprehension

```python
# Squaring numbers in a range
squared_numbers = [num ** 2 for num in numbers]
print(squared_numbers)
```

### Dictionary Comprehensions

```python
# create a dictionary with number and its square
squared_dict = {num : num ** 2 for num in numbers}
print(squared_dict)
```

### Set Comprehensions

```python
# create a set of even numbers
even_numbers = {num for num in numbers if num % 2 == 0}
print(even_numbers)
```

# Signature
Positional and Keyword Arguments

```python
def describe_pet(animal_type, pet_name):
    print(f"I have a {animal_type} named {pet_name}.")

# Call the function with positional arguments
describe_pet('dog', 'Buddy')

# Call the function with keyword arguments
describe_pet(animal_type='cat', pet_name='Whiskers')
```

# Chapter: Object-oriented Programming

### Undertanding classes and objects

```python
# basic class defination
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says Woof!")

# create an instance of the Dog class
my_dog = Dog("Buddy", 3)
my_dog.bark()
```

#### Building Custom classes

```python
#more complex custom class

class Car: 
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0

    def display_info(self):
        print(f"Car Information: {self.year} {self.make} {self.model}")

    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")

# create an instance of the Car class
my_car = Car("Toyota", "Corolla", 2020)
my_car.display_info()
my_car.update_odometer(15000)
my_car.read_odometer()
```

### Special Methods and Inheritance

Special methods allow you to define certain operations that are built into Python’s syntax. These are also called magic or dunder methods.


```python
# Example of Special methods    

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"'{self.title}' by {self.author}"

    def __len__(self):
        return len(self.title)

# create an instance of the Book class
my_book = Book("1984", "George Orwell")
print(my_book)
print(f"Length of title: {len(my_book)} characters")
```

### Inheritance: 
Inheritance allows one class to inherit the attributes and methods of another class. This helps to reuse code.

```python
# Inheritance example

class ElectricCar(Car):
    def __init__(self, make, model, year):
        super().__init__(make, model, year)
        self.battery_size = 75

    def display_battery_info(self):
        print(f"This electric car has a {self.battery_size}-kWh battery.")

# create an instance of the ElectricCar class
my_electric_car = ElectricCar("Tesla", "Model S", 2020)
my_electric_car.display_info()
my_electric_car.display_battery_info()
```
