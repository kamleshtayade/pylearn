### Iterators and Generators

#### Iterators: 
Iterators are objects that can be iterated upon. An iterator retrieves its elements by calling the function until no items are left.

```python
# example of create an interator
numbers = [1,2,3,4]
iter_obj = iter(numbers)
 
print(next(iter_obj))
print(next(iter_obj))
```

#### Generators: 
Generators are a simple way to create iterators using functions. Instead of returning a value, a generator yields a series of values.

```python
# example of generators

def countdown(num):
    print("Starting Countdown:")
    while num > 0:
        yield num
        num -= 1

for number in countdown(5):
    print(number)
```

### Decorators and Context Managers

Decorators and context managers are powerful features in Python for modifying and managing the behavior of functions and code blocks.

#### Decorators: 
A decorator is a function that takes another function and extends its behavior without explicitly modifying it.


```python
# example of decorators

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
# Another example of decorators

def debug(func):
    def wrapper(*args, **kwargs):
        print(f"Debug: Calling {func.__name__} with args: {args} and kwargs: {kwargs}")
        result = func(*args, **kwargs)
        print(f"Debug: {func.__name__} returned: {result}")
        return result
    return wrapper

@debug
def add(a, b):
    return a + b

@debug
def say_hello(name):
    return f"Hello, {name}!"

add(5, 10)
say_hello("Alice")
```

```python
# Example: using @debug with both args and kwargs
# This uses the `debug` decorator defined earlier in the notebook.

@debug
def create_order(product, quantity, *args, **kwargs):
    """Create a simple order dict. Shows how kwargs and args are passed through."""
    # Unpack some expected kwargs with defaults
    price = kwargs.get('price')
    discount = kwargs.get('discount', 0.0)
    customer = kwargs.get('customer', 'Anonymous')

    # Compute total if price is provided
    total = None
    if price is not None:
        total = (price * quantity) * (1 - discount)

    # Return a summary including any extra positional or keyword args
    return {
        'product': product,
        'quantity': quantity,
        'extra_args': args,
        'extra_kwargs': {k: v for k, v in kwargs.items() if k not in ('price','discount','customer')},
        'customer': customer,
        'price': price,
        'discount': discount,
        'total': total,
    }

# Calls demonstrating positional args + kwargs
create_order('Portable SSD 1TB', 2, 'express', price=109.99, discount=0.1, customer='Bob', gift_wrap=True)
create_order('Smartphone Tripod', 3, price=15.25, customer='Alice')
```

#### Context Managers: 
Context managers allow you to allocate and release resources precisely when you want. The most common way to create a context manager is by using the statement.

```python
# Example of Context Managers

class MyContextManager:
    def __enter__(self):
        print("Entering the context.")
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting the context.")

with MyContextManager() as manager:
    print("Inside the context.")
```

```python
# another example of context managers: This context manager handles opening and closing a file.

from contextlib import contextmanager

@contextmanager
def managed_file(name):
    print(f"Opening file: {name}")
    try:
        yield
    finally:
        print(f"Closing file: {name}")

with managed_file('example.txt'):
    print("Inside the context.")
```

#### Lambda Functions and Functional Programming

Lambda functions are small anonymous functions defined with the lambda keyword. They can have any number of arguments but only one expression.


```python
# Example of Lambda Functions

# A simple lambda function that adds 10 to the input
add_ten = lambda x: x + 10
print(add_ten(5))  # Output: 15

# A lambda function for filtering even numbers from a list
numbers = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4]

# A lambda function for double each number using map()
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)  # Output: [2, 4, 6, 8, 10]

```
