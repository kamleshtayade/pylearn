# Exception Handling

```python
result = 10/0
```

```python
try:
    result = 10/0
except ZeroDivisionError:
    print("Handle the exception - do not devided by zero")
```

```python
try:
    result = 10/0
except ZeroDivisionError:
    print("Handle the exception - do not devided by zero")
finally:
    print("execution complete - irrespective of successful/fail")
```

```python
try:
    result = 10/0
except : # no specifc error , work for all
    print("Handle the exception - do not devided by zero")
finally:
    print("execution complete - irrespective of successful/fail")
```

```python
# Example : Division by zero execption handling 

try:
    numerator = int(input("Enter the numerator: "))
    denominator = int(input("Enter the denominator:  "))

    result = numerator/denominator

    print("Result of devision : ", result)
except ZeroDivisionError:
    print("Error: division by zero is not allowed")
except ValueError:
    print("Error: Please enter valid integer values")
except Exception as e:
    print("An unexpected error occured: ", e)
finally:
    print("execution complete - irrespective of successful/fail")
```
