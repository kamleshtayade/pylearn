# All 71 Python Built-in functions explained

## Math & Numbers

### - abs()

Returns the absolute (positive) value of a number.

```python
abs(-9) # Returns the absolute value of -9
```

```python
abs(3.5) # Returns the absolute value of 3.5
```

### - divmod()

Returns both quotient and reminder as a tuple.

```python
divmod(10, 3) # (3, 1) -> 10 / 3 = 3, 10 % 3 = 1
```

### - max()

Returns the largest value from an interable or agruments. 

```python
max(3,7,2) # Returns the maximum value among the arguments - 7
```

```python
max([10, 5, 8]) # Returns the maximum value among the list - 10
```

### - min()

Returns the smallest value from an iterable or arguments.

```python
min(3, 7, 2) # Returns the minimum value among the arguments - 2
```

```python
min([10,5,8]) # Returns the minimum value among the list - 5
```

### - pow()

Raises a number to a power; optional third arg for modulo.

```python
pow(2, 8) # Returns 2 raised to the power of 8 - 256 (2 ** 8)
```

```python
pow(2,8,10) # Returns 2 raised to the power of 8, then modulo 10 - 6 (256 % 10 = 6)
```

### - round()

Rounds a number to a given number of decimal places

```python
round(3.14159, 2) # Returns 3.14
```

```python
round(3.5) # Returns 4
```

### - sum()

Adds all items in an iterable, with an optional start value.

```python
sum([1, 2, 3]) # Returns the sum of the list - 6
```

```python
sum([1,2,3],10) # Returns the sum of the list plus the second argument - 16
```

## Type Conversion

### - bin()

Converts an integer to its binary string representation

```python
bin(10) # Returns the binary representation of 10 - '0b1010'
```

```python
bin(255) # Returns the binary representation of 255 - '0b11111111'
```

### - bool()

Converts a value to True or False.

```python
bool(0) # Returns False
bool(1) # Returns True
bool("") # Returns False
bool("Hello") # Returns True
bool([]) # Returns False
bool([1, 2, 3]) # Returns True
```

### - bytearray()

Creates a mutable sequence of bytes

```python
bytearray(b"Hello") # Returns a bytearray representation of the string "Hello"
```

```python
bytearray(3) # Returns a bytearray of size 3 with all elements initialized to 0 - bytearray(b'\x00\x00\x00')
```

### - bytes()

Creates an inmutable sequence of bytes

```python
bytes(3) # Returns a bytes object of size 3 with all elements initialized to 0 - b'\x00\x00\x00'
```

```python
bytes("hi", "utf-8") # Returns a bytes object representing the string "hi" encoded in UTF-8 - b'hi'
```

### - chr()

Returns the character for a given Unicode code point. 

```python
chr(65) # Returns the character for Unicode code point 65 - 'A'
```

```python
chr(9829) # Returns the character for Unicode code point 9829 - '♥'
```

### - complex()

Creates a complex number from real and imaginary parts. 

```python
complex(2, 3) # Returns a complex number with real part 2 and imaginary part 3 - (2+3j)
```

```python
complex("4+5j") # Returns a complex number with real part 4 and imaginary part 5 - (4+5j)
```

### - float()

Converts a string or number to a floating point number.

```python
float("3.14") # Returns the floating point number 3.14
```

```python
float(5) # Returns the floating point number 5.0
```

### - hex()

Converts an integer to a hexadecimal string.

```python
hex(255) # Returns the hexadecimal representation of 255 - '0xff'
```

```python
hex(16) # Returns the hexadecimal representation of 16 - '0x10'
```

### - int()

Converts a value to an integer; optional base for strings.

```python
int("42") # Returns the integer 42
```

```python
int(3.9) # Returns the integer 3
```

```python
int("1010", 2) # Returns the integer 10, interpreting the string as a binary number
```

### - oct()

Converts an integer to its octal string representation.

```python
oct(8) # Returns the octal representation of 8 - '0o10'
```

```python
oct(64) # Returns the octal representation of 64 - '0o100'
```

### - ord()

Returns the Unicode code point of a character.

```python
ord("A") # Returns the Unicode code point for 'A' - 65
```

```python
ord('*')    # Returns the Unicode code point for '*' - 42
```

### - str()

Converts any value to its string representations

```python
str(100) # Returns the string '100'
str(3.14) # Returns the string '3.14'
str([1, 2, 3]) # Returns the string '[1, 2, 3]'
str(True) # Returns the string 'True'
```

## Data structures

### - dict()

Creates a dictionary from keyword arguments or an interable.

```python
dict(a=1, b=2) # Returns a dictionary with keys 'a' and 'b' and values 1 and 2 - {'a': 1, 'b': 2}
```

```python
dict([('x', 10), ('y', 20)]) # Returns a dictionary created from the list of key-value pairs - {'x': 10, 'y': 20}
```

### - frozenset()

Creates an immutable, unordered set of unique elements

```python
frozenset([1, 2, 2, 3]) # Returns a frozenset containing the unique elements from the list - frozenset({1, 2, 3})
# can't add/remove items unlike regular sets
```

```python
frozenset([1, 4, 3, 2, 3, 2.9, 5]) # Returns a frozenset containing the unique elements from the list - frozenset({1, 2, 3, 4, 5})
```

### - list()

Creates a mutable list from any iterable.

```python
list([1, 2, 3]) # Returns a list created from the iterable - [1, 2, 3]
```

```python
list('abc') # Returns a list of characters from the string - ['a', 'b', 'c']
```

```python
list(range(5)) # Returns a list of numbers from 0 to 4 - [0, 1, 2, 3, 4]
```

### - memoryview()

Exposes internel buffer of bytes like objects without copying. 

```python
mv = memoryview(b'hello') # Creates a memoryview object from the bytes object b'hello'
mv[0] # Accesses the first byte of the memoryview - 104 (ASCII value of 'h')
```

### - object()

Returns a plain featureless base object (base of all classes).

```python
o = object() # Returns a plain featureless base object (base of all classes)
```

### - range()

Generates a sequence of numbers (lazy, memory efficient)

```python
list(range(5)) # Returns a list of numbers from 0 to 4 - [0, 1, 2, 3, 4]
```

```python
list(range(1,10,2)) # Returns a list of numbers from 1 to 9 with a step of 2 - [1, 3, 5, 7, 9]
```

### - set()

Creates a mutable, unordered collection of unique elements.

```python
set([1, 2, 2, 3]) # Returns a set containing the unique elements from the list - {1, 2, 3}
```

### - slice()

Creates a slice object to etract parts of sequences.

```python
s = slice(1, 5) # Creates a slice object that can be used to slice sequences

[0,1,2,3,4,5][s] # Applies the slice to the list - [1, 2, 3, 4]
```

```python
s1 = slice(1, 5, 2) # Creates a slice object that can be used to slice sequences

[0,1,2,3,4,5][s1] # Applies the slice to the list - [1, 3]
```

### - tuple()

Creates an immutable ordered sequence from an iterable. 

```python
tuple([1, 2, 3]) # Converts the list to a tuple - (1, 2, 3)
```

```python
tuple("abc") # Converts the string to a tuple of characters - ('a', 'b', 'c')
```
