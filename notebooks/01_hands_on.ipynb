# Sample products dataset

This notebook includes a small sample dataset of products with prices and stock counts. Columns:

- `product`: product name
- `price`: unit price in USD
- `in_stock`: number of units currently in stock

The next cell creates a pandas DataFrame with example rows and saves it to `sample_products.csv`.

```python
import pandas as pd

# Create sample products dataset
products = [
    {"product": "Wireless Mouse", "price": 26.99, "in_stock": 150},
    {"product": "Mechanical Keyboard", "price": 89.5, "in_stock": 85},
    {"product": "USB-C Charger", "price": 19.99, "in_stock": 240},
    {"product": '27" Monitor', "price": 199.0, "in_stock": 30},
    {"product": "Laptop Stand", "price": 34.75, "in_stock": 120},
    {"product": "Noise Cancelling Headphones", "price": 129.99, "in_stock": 40},
    {"product": "Webcam 1080p", "price": 49.0, "in_stock": 95},
    {"product": "Portable SSD 1TB", "price": 109.99, "in_stock": 60},
    {"product": "Bluetooth Speaker", "price": 59.5, "in_stock": 70},
    {"product": "Smartphone Tripod", "price": 15.25, "in_stock": 200},
]


df = pd.DataFrame(products)

# Save to CSV in workspace
output_path = "./uploaded_files/sample_products1.csv"
df.to_csv(output_path, index=False)

# Display the DataFrame
df
```

```python
# Install matplotlib and draw scatter plot of price vs in-stock
# %pip install matplotlib --quiet

import matplotlib.pyplot as plt

# Ensure DataFrame `df` exists; if not, try to load the CSV
if 'df' not in globals():
    import pandas as pd
    df = pd.read_csv('./uploaded_files/sample_products.csv')

plt.figure(figsize=(10,6))
plt.scatter(df['in_stock'], df['price'], color='tab:blue', s=80, alpha=0.85, edgecolors='k')

# Annotate each point with the product name (small offset)
for i, row in df.iterrows():
    plt.annotate(row['product'], (row['in_stock'], row['price']), textcoords='offset points', xytext=(6,3), fontsize=8)

plt.xlabel('Units in stock')
plt.ylabel('Price (USD)')
plt.title('Price vs Units in Stock (sample products)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Save the plot to the repo images folder
output_img = './images/price_vs_stock.png'
plt.savefig(output_img, dpi=150)
plt.show()

print(f"Plot saved to {output_img}")
```
